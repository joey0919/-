### 고지혈증치료 결정하는 위험요인
1. 흡연
1. 고혈압(BP≥140/90 mmHg 또는 항고혈압제 복용)
1. 낮은 고밀도지단백콜레스테롤(HDL-C)(<40 mg/dL)
1. 관상동맥질환 조기 발병의 가족력(부모, 형제자매 중 남자<55세, 여자<65세에서 관상동맥질환이 발병한 경우)
1. 연령(남자≥45세, 여자≥55세)  
/* HDL-C≥60 mg/dL은 보호인자로 간주하여 총 위험요인 수에서 하나를 감한다

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
result_list = list()
```

## 2011 고지혈증 분석

```python
year = '2011'

df = pd.read_csv(f'NHIS{year}.CSV', encoding='CP949')
df.head(100)

# 1. 총콜레스테롤
filt1 = (df['총콜레스테롤'] >= 240)

# 2. HDL콜레스테롤
filt2 = (df['HDL콜레스테롤'] < 40)

#3. 중성지방
filt3 = (df['트리글리세라이드'] >= 200)

#4. LDL콜레스테롤
filt4 = (df['LDL콜레스테롤'] >= 160)

# 1개 이상의 조건 만족
filt_combined = (filt1.astype(int) + filt2.astype(int) + filt3.astype(int) + filt4.astype(int)) >= 1

# 결과
result_df = df.loc[filt_combined]
row_count = result_df.shape[0]
result_list.append(row_count)

result_df
```

## 결과: 고지혈증 증가율

```python
year_list = [2011, 2012, 2013, 2014, 2015, 2016, 2017]
result_list

plt.figure(figsize=(14, 10))
plt.plot(year_list, result_list, marker='o')

for idx, txt in enumerate(result_list):
    plt.text(year_list[idx], result_list[idx] + 100, txt, ha='center')

plt.show()
```
