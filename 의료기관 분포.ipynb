# 의료기관 분포

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

## 보건복지부 공공 보건 의료기관 지역별 분포

보건복지부_공공보건 의료기관 현황_20161231  
https://www.data.go.kr/data/15004305/fileData.do

```python
df = pd.read_csv('보건복지부_공공보건 의료기관 현황_20161231.CSV', encoding='CP949')

js = df['주소'].str.split(" ", expand=True)
grouped = js.groupby(0).size()
result = grouped.reset_index(name='Count')



result.set_index(0, inplace=True)

result.loc['서울특별시'] = result.loc['서울특별시'] + result.loc['서울시']
result.loc['경기도'] = result.loc['경기도'] + result.loc['경기']
result.loc['경상북도'] = result.loc['경상북도'] + result.loc['경북'] + result.loc['경산시']
result.loc['경상남도'] = result.loc['경상남도'] + result.loc['경남'] + result.loc['창원시']
result.loc['전라북도'] = result.loc['전라북도'] + result.loc['전북']
result.loc['전라남도'] = result.loc['전라남도'] + result.loc['전남']
result.loc['충청북도'] = result.loc['충청북도'] + result.loc['충북']
result.loc['충청남도'] = result.loc['충청남도'] + result.loc['충남'] + result.loc['천안시']
result.loc['부산광역시'] = result.loc['부산광역시'] + result.loc['부산특별시']
result.loc['대전광역시'] = result.loc['대전광역시'] + result.loc['대전시']
result.loc['강원특별자치도'] = result.loc['강원도']

result.drop(labels='서울시', axis=0, inplace=True)
result.drop(labels='경기', axis=0, inplace=True)
result.drop(labels='경북', axis=0, inplace=True)
result.drop(labels='경남', axis=0, inplace=True)
result.drop(labels='전북', axis=0, inplace=True)
result.drop(labels='전남', axis=0, inplace=True)
result.drop(labels='충북', axis=0, inplace=True)
result.drop(labels='충남', axis=0, inplace=True)
result.drop(labels='창원시', axis=0, inplace=True)
result.drop(labels='천안시', axis=0, inplace=True)
result.drop(labels='경산시', axis=0, inplace=True)
result.drop(labels='부산특별시', axis=0, inplace=True)
result.drop(labels='대전시', axis=0, inplace=True)
result.drop(labels='강원도', axis=0, inplace=True)
result

plt.figure(figsize=(15, 9))
plt.xticks(rotation=90)

plt.bar(result.index, result['Count'])
```
