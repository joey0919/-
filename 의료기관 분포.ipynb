# 의료기관 분포

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

## 보건복지부 공공 보건 의료기관 지역별 분포

```python
df = pd.read_csv('보건복지부_공공보건 의료기관 현황_20161231.CSV', encoding='CP949')

js = df['주소'].str.split(" ", expand=True)
grouped = js.groupby(0).size()
result = grouped.reset_index(name='Count')


plt.figure(figsize=(15, 9))
plt.xticks(rotation=90)

plt.bar(result[0], result['Count'])
```
